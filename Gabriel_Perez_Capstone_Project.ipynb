Capstone Project Intro

```python
# My Capstone Project Choice
MY_DATASET    = "nvb_energy_consumption.csv"
QUESTION1   = "Which building zone consumes the most energy and at what times of day?"
QUESTION2   = "Is there a relationship between occupancy level and power consumption?"
MY_DOMAIN     = "Electrical Engineering"

print(f"Dataset:  {MY_DATASET}")
print(f"Domain:   {MY_DOMAIN}")
print(f"Question #1: {QUESTION1}")
print(f"Question #2: {QUESTION2}")
```

Import Data

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://drive.google.com/uc?id=1hKMwOwPjcv_xyj105IOkmzOoDURPw1F9'
df = pd.read_csv(url)
print(f'Loaded: {df.shape[0]} rows, {df.shape[1]} columns')
```

Exploring the Data

```python
# Print the shape, column types, first 5 rows, and basic stats
print(f"Shape: {df.shape}")
print()
print("Column types:")
print(df.dtypes)
print()
print("First 5 rows:")
print(df.head())
print()
print("Basic statistics:")
print(df.describe().round(2))
```

```python
# Find missing values
print(df.isnull().sum())

# Fill columns with missing values with median
columns_to_fill = ['power_kw', 'temperature_c', 'occupancy']
for col in columns_to_fill:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)

# Verify that there are no more missing values
print(df[columns_to_fill].isnull().sum())
```

```python
# Create groupby of building zone, hour of day, and average of power kw for Question #1
zone_hourly = df.groupby(["building_zone", "hour_of_day"])["power_kw"].mean().reset_index()
print(zone_hourly)

# Keep only business hours (8 AM to 6 PM)
business_hours_df = df[(df["hour_of_day"] >= 8) & (df["hour_of_day"] <= 18)]

# Keep only rows where occupancy is greater than zero
occupied_df = df[df["occupancy"] > 0]
```

Graphs

```python
# Graph for Question #1
import matplotlib.pyplot as plt

# Set up a clean, readable chart style
plt.figure(figsize=(12, 6))

# Iterate through each building zone to plot its energy consumption
for zone in zone_hourly["building_zone"].unique():
    zone_data = zone_hourly[zone_hourly["building_zone"] == zone]
    plt.plot(zone_data["hour_of_day"], zone_data["power_kw"], marker='o', label=zone)

# Label the chart clearly
plt.title("Daily Energy Consumption Patterns by Building Zone", fontsize=14)
plt.xlabel("Hour of the Day (0-23)", fontsize=12)
plt.ylabel("Average Energy Consumption", fontsize=12)
plt.xticks(range(0, 24))  # Show every hour on the X-axis
plt.legend(title="Building Zone") # Add a legend

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Graph for question #2
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
# Create a scatter plot
plt.scatter(df['occupancy'], df['power_kw'], alpha=0.5)
plt.title('Occupancy level vs Power Consumption', fontsize=14, fontweight='bold')
plt.xlabel('Number of People')
plt.ylabel('Power Consumption (kW)l')

plt.tight_layout()
plt.show()
```
